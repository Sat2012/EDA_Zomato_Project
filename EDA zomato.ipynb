```python
country_df=pd.read_excel('Country-Code.xlsx')
```

```python
country_df.shape
```

```python
final_df=pd.merge(df,country_df, on='Country Code', how='left')
```

```python
final_df.head(5)
```

```python
final_df.shape
```

```python
final_df.info()
```

```python
# finding the countries with the most ordered
final_df.Country.value_counts()
```

```python
country_names= final_df.Country.value_counts().index
values= final_df.Country.value_counts().values
```

```python
## making the pie chart for the top 5 countries which is using zomato more
plt.pie(values[:3],labels=country_names[:3], autopct='%1.2f%%')
```

# Observation 1
There are most of the transaction or used in India followed by the US and UK since it Zomaro have major Business that is based in India

```python
#Exploring more about the data
final_df.columns
```

```python
rating_count=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
```

```python
rating_count
```

## Observations
1. When there is rating is between 04.5 to 4.9 are Excellent
2. When there is rating is between 04.0 to 4.4 are Very Good
3. When there is rating is between 03.5 to 3.9 are Good
4. When there is rating is between 02.5 to 3.4 are Average
5. When there is rating is between 1.8 to 2.4 are Poor
6. many people does not rate the zomato at all 

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(15,6)
sns.barplot(x='Aggregate rating', y='Rating Counts', hue='Rating color',data=rating_count,palette=['blue','red','orange','yellow','green','green'])
```
